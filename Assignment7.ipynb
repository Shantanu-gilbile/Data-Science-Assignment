```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('titanic_train.csv')
```

```python
data.head()
```

```python
data['sex']
```

```python
data['sex'].value_counts().plot(kind='bar')
```

```python
sns.histplot(data['age'],kde=True,bins=20)
```

```python
data['age'].plot(kind='kde')
```

```python
sns.countplot(x ="survived", hue ="sex", data = data)
```

```python
plt.pie(data['survived'],labels=data['survived'],autopct="%.0f%%")
plt.show()
```

Female survival Rate is more tha Male

```python
survived_data = data[data['survived'] == 1]
survived = survived_data.count().values[1]
survival_percent = (survived/850) * 100
print('The percentage of survived people in training data are {}'.format(survival_percent))
```

```python

data['Fare_Range'] = pd.qcut(data['fare'], 4)
sns.barplot(x ='Fare_Range', y ='survived',data = data)
```

High Fare has more Survival rate

```python
q1 = np.percentile(data,25)
q3 = np.percentile(data,75)
q1,q3
```

```python
IQR = q3 - q1
```

```python
lower = q1 - 1.5*(IQR)
Upper = q3 + 1.5*(IQR)

print(f"Range is : ({lower},{Upper})")
```

```python
outliers = []
for i in data:
    if i < lower or i > Upper:
        outliers.append(i)

outliers
```

```python
def Outliers(col):
    
    q1 = np.percentile(col,25)
    q3 = np.percentile(col,75)
    IQR = q3 - q1

    lower = q1 - 1.5*(IQR)
    Upper = q3 + 1.5*(IQR)

    print(f"Range is : ({lower},{Upper})")

    outliers = []
    
    for i in col:
        if i < lower or i > Upper:
            outliers.append(i)

    return outliers
```

```python
def MinMaxScaler(col):
    
    min = np.min(col)
    max = np.max(col)

    Scaled = []
    for i in col:
        scaled = np.round((i - min)/(max - min),6)
        Scaled.append(scaled)

    return Scaled
```
